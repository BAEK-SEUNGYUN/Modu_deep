### image 저장하기: cv2.imwrite()

```python
# cv2.imwrite(filename, img, params =None) -> retval


# filename: 저장할 파일 이름
# img: 저장할 자료 (numpy.ndarray)
# params: 파일저장 옵션
# retval: 성공 True, 실패 False
```

### 새창띄우기 : cv2.namedWindow()

```python
# cv2.namedWindow(winname, flags =None) -> None

# winname: 창 이름
# flags:
# cv2.WINDOW_NORMAL: 영상크기를 창 크기에 맞게 지정
# cv2.WINDOW_AUTOSIZE: 창크기를 영상 크기에 맞게 변경
```

### image 저장하기: cv2.imshow()

```python
# imshow(winname, mat) -> None?

# winname: 창 이름
# mat: numpy.ndarray (default = uint8)
# uint16, int32의 경우 255로 나누어서 출력
# float32, float64의 경우 0 ~1로 만든 후 행렬값에 255를 곱해서 출력
```

### 키보드 입력대기: cv2.waitKey()

```python
# waitKey([, delay]) -> retval
# delay: Delay in milliseconds. 0 is the special value that means "forever"

# retval
```

```python
# img = cv2.imread('cat.bmp')
img = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE)
# img = cv2.imread('cat.bmp', cv2.IMREAD_UNCHANGED)


if img is None:
    print('Image load failed')
    sys.exit()


######
# cv2.imwrite('cat_gray.png', img)
cv2.imwrite('cat_gray.jpeg', img)
    
cv2.namedWindow('Hello OpenCV', cv2.WINDOW_AUTOSIZE)
# cv2.namedWindow('Hello OpenCV', cv2.WINDOW_NORMAL)

cv2.imshow('Hello OpenCV', img)

# Keyboard 입력 (모든 키) 을 기다리는 동안 영상을 출력
# key = cv2.waitKey(), key =  -1, or ASCII

# cv2.waitKey()
# while True:
#     if cv2.waitKey() == 27: # esc 의 아스키 코드
#         break

while True:
    if cv2.waitKey() == ord('q'):# q를 누르면 종료
        break
    
 
# 키보드 입력을 받으면 화면에 나타난 모든 창을 닫음
cv2.destroyAllWindows()
```

### matplotlib 라이브러리를 이용한 영상출력

```python
import matplotlib.pyplot as plt
```

```python
# img = cv2.imread('cat.bmp')
imgBGR = cv2.imread('cat.bmp', cv2.IMREAD_COLOR)
# img = cv2.imread('cat.bmp', cv2.IMREAD_UNCHANGED)


if img is None:
    print('Image load failed')
    sys.exit()


# BGR to RGB
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

## plt를 이용한 영상출력
plt.subplot(121), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(122), plt.imshow(imgRGB), plt.axis('off')
plt.show()
```

```python
## gray scale 영상으로 출력
imgGray = cv2.imread('cat.bmp', cv2.IMREAD_GRAYSCALE)

# plt.imshow(imgGray, cmap='gray')
# plt.axis('off')
# plt.show()

## 두 개의 영상을 모두 출력
plt.subplot(121), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(122), plt.imshow(imgGray, cmap = 'gray'), plt.axis('off')
plt.show()
```

## 이미지 슬라이드쇼

```python

import sys
## Wild card
import glob
import cv2

img_files = glob.glob('./images/*.jpg')

for f in img_files:
    print(f)
    
# 전체 화면으로 'image' 창 생성
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

cnt = len(img_files) # len = 5
idx = 0

while True:
    img = cv2.imread(img_files[idx])
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(1000) == 27: #ESC
        break
    
    idx += 1
    if idx >= cnt:
        idx = 0
        
cv2.destroyAllWindows()
```
