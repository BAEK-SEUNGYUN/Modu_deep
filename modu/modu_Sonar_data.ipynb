```python
import tensorflow as tf
import pandas as pd
import numpy as np

from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
# seed 값 설정
seed = 0
np.random.seed(3)
tf.compat.v1.set_random_seed(3)
```

```python
# 데이터 적용
df = pd.read_csv(my_data, header=None)
df
```

```python
df.info()
```

```python
df.head()
```

```python
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]
print(X)
print(Y_obj)
```

```python
# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

print(Y_obj)
print(Y)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(24,input_dim=60, activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# 모델 컴파일
model.compile(loss = 'mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X,Y, epochs=200, batch_size=5)
```

```python
print('Accuracy : %.4f' % model.evaluate(X,Y)[1])
```

과적합 : 모델이 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것

과적합을 방지하려면 -> 학습 데이터셋, 테스트 데이터셋으로 분리

학습과 동시에 테스트를 병행하며 진행하는 것이 한 방법


x데이터와 y 데이터에서 각각 정해진 비율만큼 구분하여 한 그룹은 학습에 사용하고 다른 한 그룹은 테스트에 사용하게 하는 함수가 sklearn 라이브러리의 **train_test_split()** 함수

```python
from sklearn.model_selection import train_test_split
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.compat.v1.set_random_seed(3)
```

```python
#데이터 적용
df = pd.read_csv(my_data, header=None)
```

```python
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 학습셋과 테스트셋의 구분

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=130, batch_size=5)
```

```python
# 테스트셋에 모델적용

print("\n Test Accuracy : %.4f" %(model.evaluate(X_test, Y_test)[1]))
```
